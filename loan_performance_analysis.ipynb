## Dataset Overview

The dataset consists of loan information, including loan amount, grade, term, and various other features.
Below are the key steps involved:
- **Loading the Dataset**: The dataset is loaded and initial structure is explored.
- **Preprocessing**: Missing values are handled, and categorical data is encoded.


```python

# Load Dataset
import pandas as pd

data = pd.read_csv('lending_club_loan_two.csv')
data.head()
```

## Data Preprocessing

Steps included in preprocessing:
1. Converted the `term` column to numeric.
2. Handled missing values in numeric columns.
3. Encoded categorical features with low cardinality using one-hot encoding.
4. Removed high-cardinality and irrelevant columns.


```python

# Preprocessing Steps (Simplified Example)
data['term'] = data['term'].str.replace(' months', '').astype(int)
data = pd.get_dummies(data, drop_first=True)
data.head()
```


## Model Training and Evaluation

A Random Forest Classifier was trained on the processed data. The following metrics were used for evaluation:
- **Precision**: The ability of the model to correctly identify positive instances.
- **Recall**: The ability of the model to capture all relevant positive instances.
- **F1-Score**: A weighted average of precision and recall.


## Insights

### Key Findings:
1. Loans in higher grades (e.g., F, G) have significantly higher average loan amounts.
2. "Debt Consolidation" is the most common loan purpose, as seen in the word cloud.
3. The model performs well for identifying non-default loans (class 0) but struggles with default loans (class 1).

### Recommendations:
- Further feature engineering may improve model performance for default prediction.
- Investigate additional features or external datasets to capture borrower behavior more effectively.



## Issues Faced and Solutions

1. **Memory Issue with High-Cardinality Columns**:
   - Columns like `emp_title` and `desc` had too many unique values, causing memory issues during one-hot encoding.
   - **Solution**: Dropped high-cardinality columns.

2. **Non-Numeric Columns in Model Input**:
   - Columns like `term` were non-numeric and caused errors during model training.
   - **Solution**: Converted such columns to numeric formats.

3. **Model Performance on Default Loans**:
   - The model struggled to predict default loans (class 1) accurately due to class imbalance.
   - **Solution**: Suggested further feature engineering and oversampling techniques to address class imbalance.



## What I Learned

1. Effective data preprocessing is critical for machine learning pipelines.
2. Exploratory Data Analysis helps uncover key insights into the dataset, guiding feature engineering.
3. Random Forest performs well for this dataset but struggles with imbalanced classes, highlighting the importance of class balancing techniques.
4. Proper error handling and memory management are essential for working with large datasets.
